```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# simple example

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#plotting using matplotlib

#plt scaatter

plt.scatter(x,y,c='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
plt.savefig("scatter.png")
```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'r*--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
plt.savefig("plot.png")
```

```python
#creating subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go--')
```

```python
x=np.arange(1,11)
y= 3*x+5
plt.title('Matplotlib demo')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
## sine curve

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')

#plot points using matplotlib

plt.plot(x,y)
plt.show()
```

```python
## sine and cosine curves

x= np.arange(0, 5 * np.pi, 0.1)
y_sin= np.sin(x)
y_cos= np.cos(x)

#use subplot with height=2 and width=1
plt.subplot(2, 1, 1)

#first plot

plt.plot(x,y_sin,'r--')
plt.title('Sine')

#second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

#show plot
plt.show()

```

```python
## Bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y,align= 'center')
plt.bar(x2,y2,color='g')
plt.title("Bar graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
## Histogram

a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
plt.title(" Histogram")
plt.show()
```

```python
## Box plot using matplotlib

data=[np.random.normal(0, std, 100) for std in range(1, 4)]
# plt box plot

plt.boxplot(data, vert= True, patch_artist = True);
```

```python
## Pie chart

labels='python','c++','Ruby','java'
sizes=[215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #1st slice

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
